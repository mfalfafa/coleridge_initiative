```python
import urllib.request, json 
import pandas as pd
#from pandas.io.json import json_normalize
############## https://labs.data.gov/dashboard/offices/qa
api_list = ["http://www.usda.gov/data.json","http://www.commerce.gov/data.json","http://www.ed.gov/data.json","https://healthdata.gov/data.json","http://www.dol.gov/data.json","https://www.doi.gov/sites/doi.gov/files/data.json","http://www.energy.gov/data.json","http://www.dhs.gov/data.json","http://www.ustreas.gov/data.json","http://www.dot.gov/data.json","http://www.va.gov/data.json","http://www.gsa.gov/data.json","http://www.nsf.gov/data.json","http://www.sba.gov/data.json","http://www.ssa.gov/data.json","http://www.usaid.gov/data.json"]
#api_list = ["http://www.commerce.gov/data.json"]
i=0
for site in api_list:
    print (site)
    with urllib.request.urlopen(site) as url:
        data = json.loads(url.read().decode())
        #print (data)
        if i==0:
            df = pd.DataFrame(pd.json_normalize(data, 'dataset'))
        else:
            temp = pd.DataFrame(pd.json_normalize(data, 'dataset'))
            df = df.append(temp, ignore_index=True)
    i=i+1
#df=df[df['@type'].str.contains('dcat:Dataset')]
df
```
